```python
#CREAMOS LA VENTANA LOGIN
# create the input fields and login button
username_input = widgets.Text(description='',
                              placeholder='User',
                              value = '',
                              disabled=False

                              )


password_input = widgets.Password(
    description='',
    value='',
    placeholder='Password',
    disabled=False)


login_button = widgets.Button(description='Login')
# create the output widget
login_output = widgets.Output()

# create the login window
login_window = widgets.VBox([username_input, password_input, login_button, login_output])

login_window.layout = widgets.Layout(
    display='flex',
    align_items='center',
    justify_content='center'

    )
# display the login window



#DEFINIMOS QUE HACE EL BOTON LOGIN
# define the event handler for the login button
def on_login_button_clicked(b):
    with login_output:

        #login_output.clear_output()
        # handle the login functionality here
        

        try:
            datos_usuario = list(claves.find({"usuario": username_input.value}))[0]
            if datos_usuario["clave"] == password_input.value:

                # clear the input fields and clear the output widget
                username_input.value = ''
                password_input.value = ''

                #AQUI LO QUE HACEMOS ES CONVERTIR A login_output en vacío, luego entonces al ejecutar "login_window.children = [login_output]", la ventana se vuelve blanca
                #y ya podemos poder lo que queramos CREO

                login_output.clear_output()
                # remove the login widgets from the first window and create the accept window
                login_window.children = [login_output]

                # display the accept window
                display(work_zone_window)

                #establecemos el saludo
                hora_saludo = datetime.now().hour
                if 6 < hora_saludo < 12:
                    saludo = "Buenos días"
                elif 12 <= hora_saludo < 19:
                    saludo = "Buenas tardes"
                else:
                    saludo = "Buenas noches"
                habla(f"{saludo} {datos_usuario['nombre']}, Atenea analytics te da la bienvenida. Por favor, selecciona un área de trabajo.")

            else:
                # clear the input fields and show an error message
                username_input.value = ''
                password_input.value = ''
                login_window.children = [incorrect_output, retry_window]
        except:
            # clear the input fields and show an error message
            username_input.value = ''
            password_input.value = ''
            login_window.children = [incorrect_output, retry_window]
            

#CREAMOS QUE ACCION TOMA EL BOTON LOGIN
# attach the event handler to the login button
login_button.on_click(on_login_button_clicked)
```
